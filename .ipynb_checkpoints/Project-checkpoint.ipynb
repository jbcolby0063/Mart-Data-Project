```python
cust_df.dropna(inplace=True)
cust_df.isna().sum()
```

```python
# 출생 연도를 나이로 변경 (2023년 1월 기준)
cust_df['birth_year'] = 2023 - cust_df['birth_year']
cust_df.rename(columns={'birth_year': 'age'}, inplace=True)
cust_df
```

```python
# 구매 금액 합계 넣어주기

cust_df['amount_total'] = cust_df['amount_alcohol'] + cust_df['amount_fruit'] + cust_df['amount_meat'] + cust_df['amount_fish'] + cust_df['amount_snack'] + cust_df['amount_general']
# 'amount_total'을 'amount_general' 바로 옆에 배치
amount_general_index = cust_df.columns.get_loc('amount_general') # get_loc() : 해당 인덱스 출력
amount_total_data = cust_df.pop('amount_total') # pop() : column을 DataFrame에서 pop
cust_df.insert(loc=amount_general_index + 1, column='amount_total', value=amount_total_data) # insert() : 원하는 위치에 column 삽입

cust_df.head()
```

```python
# 구매 횟수 합계 넣어주기

cust_df['num_purchase_total'] = cust_df['num_purchase_web'] + cust_df['num_purchase_store'] + cust_df['num_purchase_discount']
# 'num_purchase_total'을 'num_purchase_discount' 바로 옆에 배치
num_purchase_discount_index = cust_df.columns.get_loc('num_purchase_discount') # get_loc() : 해당 인덱스 출력
num_purchase_total_data = cust_df.pop('num_purchase_total') # pop() : column을 DataFrame에서 pop
cust_df.insert(loc=num_purchase_discount_index + 1, column='num_purchase_total', value=num_purchase_total_data) # insert() : 원하는 위치에 column 삽입

cust_df.head()
```

```python
# 불필요한 column 제거

# 'ID' column은 고객 특성을 분석하는거와 연관 x -> 삭제해도 상관 없음
# 'revenue' column은 보면 평균이 11이고 std가 0이니 모든 값이 11이라는 뜻 -> 분석하는데 필요 없음
cust_df.drop(['ID', 'revenue'], axis='columns', inplace=True)
cust_df.head()
```

```python
# 구간화 (Binning) 시키기
# 나이 그래프 보기 편하게 10대, 20대, 30대 이런식으로 연령대로 나누기

# 구간 만들어주기 
age_bins = list(range(10, 81, 10))
print(f'age_bins: {age_bins}')
# 각 구간 이름 지정
age_labels = [f'{x}대' for x in age_bins[:len(age_bins) - 1]]
print(f'age_labels: {age_labels}')
# cut(x=구간화할 데이터, bins=구간 범위들, labels=각 구간 이름, right=구간의 오른쪽 값을 포함할지 안할지) : 사용자가 지정한 구간 범위대로 데이터를 나눠주기 
age_group = pd.cut(x=cust_df['age'], bins=age_bins, labels=age_labels, right=False) # 구간 오른쪽 '미만'이기 때문에 right=False
# 구간화한 데이터 cust_df에 넣어주기
cust_df.insert(loc=cust_df.columns.get_loc('age')+1, column='age_group', value=age_group)
```

```python
# 구간화한 데이터 age_group 모습
cust_df.head()
```

```python
cust_df['age_group'].value_counts()

# 70대와 10대의 숫자가 너무 적어보임. 이러면 해당 연령대에 대한 통계 수치 신뢰성이 떨어짐. 그래서 10대와 20대를 합치고 60대와 70대를 합치겠음
```

```python
# replace(d): d에 쓰여있는대로 값 변경
age_dict = {
    '10대': '20대 이하',
    '20대': '20대 이하',
    '60대': '60대 이상',
    '70대': '60대 이상'
}
cust_df['age_group'] = cust_df['age_group'].replace(age_dict)
```

```python
Q1 = cust_df['annual_income'].quantile(0.25)
Q3 = cust_df['annual_income'].quantile(0.75)
IQR = Q3 - Q1
outlier_condition = (cust_df['annual_income'] < Q1 - 1.5 * IQR) | (cust_df['annual_income'] > Q3 + 1.5 * IQR)
cust_df.drop(cust_df[outlier_condition].index, axis='index', inplace=True)

sns.boxplot(data=cust_df, y='annual_income')

# 소득 이상값 제거 확인
```

```python
# RFM 등급 매기기 (3개의 등급, 숫자가 클수록 긍정적, 3등급이 제일 좋음)

num_grades = 3 # 등급 수 
grade_labels = list(range(1, num_grades + 1)) # 각 등급 이름 : [1, 2, 3]

# 1. Recency 등급 매기기 ('recency' column값이 작은 고객일수록 최근에 구매 -> 등급 높게 부여해야함)
# qcut(x=구간화할 데이터, q=해당값만큼 구간화하여 데이터 수를 균등하게 등분, labels=각 구간 이름) : 사용자가 지정한 수만큼 구간화 시키고 각 구간마다 데이터 수가 똑같도록 분배
# cut()은 각 구간들을 똑같은 크기의 범위만큼 나누지만 qcut()은 각 구간들이 데이터 수를 똑같이 갖고 있도록 나눔 (https://stackoverflow.com/questions/30211923/what-is-the-difference-between-pandas-qcut-and-pandas-cut) 
recency_grade = pd.qcut(x=cust_df['recency'], q=num_grades, labels=grade_labels[::-1])
cust_df['recency_grade'] = recency_grade

cust_df['recency_grade'].value_counts()
# 1등급부터 3등급까지 비슷한 비율로 나뉜걸 볼 수 있음
```

```python
# Recency 등급마다 매출 기여도 확인
group_recency_grade = cust_df.groupby('recency_grade').sum(numeric_only=True)
group_recency_grade
```

```python
# Recency 등급마다 매출 기여도 시각화
group_recency_grade.plot(kind='pie', y='amount_total', labels=[f'{x}등급' for x in group_recency_grade.index], autopct='%.1f%%')
plt.title("Recency 등급별 매출 기여도")
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
```

```python
# 2. Frequency 등급 매기기 ('num_purchase_total' column값이 큰 고객일수록 더 자주 구매 -> 등급 높게 부여해야함)
cust_df['frequency_grade'] = pd.qcut(x=cust_df['num_purchase_total'], q=num_grades, labels=grade_labels)

cust_df['frequency_grade'].value_counts()
# 1등급이 다른 등급들보다 균등하게 나눠지지 못한거 같음
```

```python
# Frequency 등급마다 매출 기여도 확인
group_frequency_grade = cust_df.groupby('frequency_grade').sum(numeric_only=True)
group_frequency_grade
```

```python
# Frequency 등급마다 매출 기여도 시각화
group_frequency_grade.plot(kind='pie', y='amount_total', labels=[f'{x}등급' for x in group_frequency_grade.index], autopct='%.1f%%')
plt.title('Frequency 등급별 매출 기여도')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
```

```python
# 3. Monetary 등급 매기기 ('amount_total' column값이 큰 고객일수록 더 많이 구매 -> 등급 높게 부여해야함)
cust_df['monetary_grade'] = pd.qcut(x=cust_df['amount_total'], q=num_grades, labels=grade_labels)
cust_df['monetary_grade'].value_counts()
# 등급별로 균등하게 분배됨
```

```python
# Monetary 등급마다 매출 기여도 확인
group_monetary_grade = cust_df.groupby('monetary_grade').sum(numeric_only=True)
group_monetary_grade
```

```python
# Monetary 등급마다 매출 기여도 시각화
group_monetary_grade.plot(kind='pie', y='amount_total', labels=[f'{x}등급' for x in group_monetary_grade.index], autopct='%.1f%%')
plt.title('Monetary 등급별 매출 기여도')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
```

```python
# weight를 이용해 RFM 고객 세그먼트 분류하기

# R,F,M 등급에 곱해줄 가중치(weight) 설정
weight ={ 
    'recency': 1/3,
    'frequency': 1/3,
    'monetary': 1/3
}
# weight이용해 각 고객들 고객 지표 계산
# 'recency_grade', 'frequency_grade', 'monetary_grade'에 있는 등급들은 숫자형 데이터가 아니라 범주형 데이터라서 astype()를 이용해 숫자형 데이터로 바꿔줘야함
cust_df['rfm_score'] = (
    weight['recency'] * cust_df['recency_grade'].astype('int') 
    + weight['frequency'] * cust_df['frequency_grade'].astype('int')
    + weight['monetary'] * cust_df['monetary_grade'].astype('int')
)

cust_df.head()
```

```python
# 고객 지표는 최소값이 1이고 최대값이 3임. 
# 그래서 구간마다 동일한 수의 고객을 분배하기보다는 최소값과 최대값을 이용하여 동일한 크기의 범위로 구간을 나누는게 좋아보임
# 1이상 5/3미만 -> 1등급, 5/3이상 7/3이하 -> 2등급, 7/3초과 3이하 -> 3등급 이렇게
# 하지만 이렇게 구간 범위에서 오른쪽 포함 여부가 일관되지 않은 경우(5/3미만, 7/3이하, ...)에는 cut()함수를 쓰기 어려움
# 그래서 apply()함수 이용
def rfm_score_bins(x):
    if x < 5/3: # 1이상 5/3미만 -> 1등급
        return 1
    elif x <= 7/3: # 5/3이상 7/3이하 -> 2등급
        return 2
    else: # 7/3초과 3이하 -> 3등급
        return 3

cust_df['rfm_segment'] = cust_df['rfm_score'].apply(rfm_score_bins)
cust_df['rfm_segment'].value_counts()
```

```python
# 고객 RFM 등급별 매출 기여도 시각화
group_rfm_segment = cust_df.groupby('rfm_segment').sum(numeric_only=True)
group_rfm_segment.plot(kind='pie', y='amount_total', labels=[f'{x}등급' for x in group_rfm_segment.index], autopct='%.1f%%')
plt.title('RFM 고객 세그먼트별 매출 기여도')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
```

```python
# 앞선 분석에서 Recency 등급은 등급별로 매출 기여도가 다 비슷비슷했음. 
# 따라서 Recency 가중치를 줄이고 다른 Frequency와 Monetary를 높이면 고객 세그멘트별 매출 기여도가 더욱 뚜렷해질 것 같음

weight['recency'] = 0.2
weight['frequency'] = 0.4
weight['monetary'] = 0.4

# weight이용해 각 고객들 고객 지표 재계산
cust_df['rfm_score'] = (
    weight['recency'] * cust_df['recency_grade'].astype('int') 
    + weight['frequency'] * cust_df['frequency_grade'].astype('int')
    + weight['monetary'] * cust_df['monetary_grade'].astype('int')
)
# RFM 지표 재구간화
cust_df['rfm_segment'] = cust_df['rfm_score'].apply(rfm_score_bins)

cust_df['rfm_segment'].value_counts()
```

```python
# 고객 RFM 등급별 매출 기여도 시각화
group_rfm_segment = cust_df.groupby('rfm_segment').sum(numeric_only=True)
group_rfm_segment.plot(kind='pie', y='amount_total', labels=[f'{x}등급' for x in group_rfm_segment.index], autopct='%.1f%%')
plt.title('RFM 고객 세그먼트별 매출 기여도')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))

# 3등급 세그먼트의 매출 기여도가 거의 70%임. 따라서 3등급 세그먼트들을 집중적으로 관리해주면 앞으로 매출에 도움이 될 것 같음.
```

```python
# 세그먼트 품목별 매출 기여도 확인
group_rfm_segment = cust_df.groupby('rfm_segment').sum(numeric_only=True).reset_index()
group_rfm_segment
```

```python
# 세그먼트 품목별 매출 기여도 시각화
condition = (group_rfm_segment.columns.str.startswith('amount_')) & (group_rfm_segment.columns != 'amount_total')
amount_columns = list(group_rfm_segment.columns[condition])
amount_columns.append('rfm_segment')
amount_items = group_rfm_segment[amount_columns]

amount_items.set_index('rfm_segment', inplace=True)
amount_items
```

```python
# 세그먼트 품목별 매출 기여도 시각화
for grade in amount_items.index:
    amount_items.loc[grade].plot(kind='pie', labels=['주류', '과일', '육류', '수산물', '과자', '잡화'], ylabel='', autopct='%.1f%%')
    plt.title(f'{grade}등급 세그먼트 품목별 매출 기여도')
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.show()

# 모든 고객 세그먼트에서 주류 매출이 가장 높음. 그다음으로 육류의 매출이 높음.
# 1등급에서 3등급갈수록 주류의 매출이 높아지지만 수산물,과일,과자,잡화는 줄어들고 있음. 부양 자녀수가 줄어드니 주류 소비가 늘고 과일,과자,잡화같은건 소비가 줄어드는 경향이 있음. 
# 3등급은 주로 주류와 육류에 집중되어있음. 이걸 이용해서 마케팅을 전개하면 좋을거 같음.
```

```python
# 세그먼트별 포로모션 참여율 확인
group_rfm_segment = cust_df.groupby('rfm_segment').mean(numeric_only=True).reset_index() # 참여율이니 mean()함수 이용
group_rfm_segment
```

```python
# 세그먼트별 포로모션 참여율 시각화
condition = (group_rfm_segment.columns.str.startswith('promotion_'))
promotion_columns = list(group_rfm_segment.columns[condition])
promotion_columns.append('rfm_segment')
promotions = group_rfm_segment[promotion_columns]

promotions.set_index('rfm_segment', inplace=True)
promotions
```

```python
# 세그먼트별 포로모션 참여율 시각화
promotions.plot(kind='bar')
plt.title("고객 세그먼트별 프로모션 참여율")
plt.xlabel('고객 세그먼트')
plt.ylabel('프로모션 참여율')
plt.grid()

# 1등급에서 3등급 갈수록 포로모션 참여율이 높아짐. 
# 3번 프로모션은 모든 고객 세그먼트에서 비슷한 참여율을 보임. 
# 2번 프로모션은 참여율이 저조했음.
# 6번 프로모션은 다른 프로모션에 비해 참여율이 높았음. 
```
